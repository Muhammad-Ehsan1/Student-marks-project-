```python
##Import Libraries
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
##Import dataset
df=pd.read_csv(r"F:\Data Analytics\Data Sources\students_scores.csv")
print(df.head())
```

```python
## Data Summary
df.describe()
```

```python
## Data Type
df.info()
```

```python
##Null Count
df.isnull().sum()
```

```python
##drop unnamed column
df=df.drop("Unnamed: 0", axis = 1)
print(df.head())
```

```python
##Gender Distribution 
plt.figure(figsize=(5,5))
ax=sns.countplot(data=df,x="Gender",hue = 'Gender')
for container in ax.containers:
    ax.bar_label(container)
plt.title("Gender Distribution")
##ax.bar_label(ax.containers[0])   (agar hue nhi dia hota to is sy de dety)
plt.show()
```

from the above chart we have analysed that :
That no of females in the data is more than no of Males

```python
## group_by by parent Education 
gb=df.groupby("ParentEduc").agg({'MathScore':'mean','ReadingScore':'mean','WritingScore':'mean'})
print(gb)
```

```python
##creating heatmap of parent education(gb)
plt.figure(figsize=(4,4))
sns.heatmap(gb,annot=True)
plt.title("Relationship b/w parents education & student marks")
plt.show()
```

From the above heatmap we can Conclude : 
That the education of the parents have a good impact on Student marks 

```python
## group_by by ParentMaritalStatus 
gb1=df.groupby("ParentMaritalStatus").agg({'MathScore':'mean','ReadingScore':'mean','WritingScore':'mean'})
print(gb1)
```

```python
##creating heatmap of ParentMaritalStatus(gb1)
plt.figure(figsize=(4,4))
sns.heatmap(gb1,annot=True)
plt.title("Relationship b/w parents martial status & student marks")
plt.show()
```

From the above heatmap we can Conclude : 
That there is no/negligible impact on student marks due to their parent's marital status

```python
## Checking outliers in cource marks
sns.boxplot(data = df,x="MathScore")
plt.show()
sns.boxplot(data = df,x="ReadingScore")
plt.show()
sns.boxplot(data = df,x="WritingScore")
plt.show()
```

from the above boxplots : 
we see their are many outliers in the score of cource

```python
##Distribution of EthnicGroup with pie chart
groupA=df.loc[(df["EthnicGroup"]=="group A")].count()
groupB=df.loc[(df["EthnicGroup"]=="group B")].count()
groupC=df.loc[(df["EthnicGroup"]=="group C")].count()
groupD=df.loc[(df["EthnicGroup"]=="group D")].count()
groupE=df.loc[(df["EthnicGroup"]=="group E")].count()

mlist=[groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
l=["group A","group B","group C","group D","group E"]
plt.figure(figsize=(4,4))
plt.pie(mlist,labels=l,autopct="%1.2f%%")
plt.title("Distribution by EthnicGroup")
plt.show()
```
